Bu iki sınıfı tek bir doğruyla ayıramazsın (lineer ayrılabilir değil). O yüzden gizli katman + doğrusal olmayan aktivasyon (ReLU/tanh) gerekir. Ağ, uzayı büküp iki bölgeyi ayırmayı öğrenir.

numpy: veriyi dizi halinde tutmak.
Sequential: Katmanları sırayla dizeriz (en basit model tipi).
Dense: Tam bağlantılı katman (her nöron, önceki katmanın tüm nöronlarına bağlı).

Mimari:

Gizli katman: Dense(5, input_dim=2, activation='relu')
input_dim=2: her girdide 2 özellik var (XOR’un iki biti).
5: 5 gizli nöron (kapasite). XOR için 2–5 arası yeterli.
relu(z)=max(0,z): doğrusal olmayanlık, gradyan kararlılığı.
Çıkış katmanı: Dense(1, activation='sigmoid')
1 nöron → ikili sınıflama.
sigmoid çıktıyı olasılığa çevirir (0–1 arası).

Parametre sayısı

Katman1: Ağırlıklar (2×5=10) + bias (5) = 15
Katman2: Ağırlıklar (5×1=5) + bias (1) = 6
Toplam = 21 öğrenilebilir parametre.

loss = binary_crossentropy (ikili sınıflamada standart):
optimizer = adam: gradyanları adaptif ölçekleyip hızlı ve stabil öğrenir.
metrics=['accuracy']: eğitim sırasında doğruluk raporla.

Veriyi 100 kez dolaş (epoch).
Bu veri küçük olduğu için aslında full-batch gidiyor (tamamını bir kerede).

predict olasılık döndürür (0–1).

> 0.5 ile sınıfa çeviriyoruz (0/1). XOR’da beklenen çıktı sırasıyla [0,1,1,0].
Not: Bazen 100 epoch yeterli olmazsa 300–1000’e çıkar; ya da relu yerine tanh da iyi çalışır. Reprodüksiyon için:

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# giriş verileri (XOR problemi)
X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
y = np.array([[0],[1],[1],[0]], dtype=np.float32)

# basit model
model = Sequential()
model.add(Dense(5, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# modeli derle
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# modeli eğitelim
model.fit(X, y, epochs=100, verbose=1)

# tahminleme
tahminleme = model.predict(X)

print('Tahminler:\n', tahminleme)
print('Yuvarlanmış sınıf tahminleri:\n', (tahminleme > 0.5).astype(int))
```
