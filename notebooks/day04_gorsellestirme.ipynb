```python
import pandas as pd
import matplotlib.pyplot as plt
```

## ÖRNEK VERİ

```python
df=pd.read_excel("teknolojik_urunler_zamanli.xlsx")
df["Tarih"]=pd.to_datetime(df["Tarih"])
df.set_index("Tarih",inplace=True)
```

## Satışların zaman içindeki değişimini gösteren çizgi grafik

```python
df['Satış'].plot(title="Satışların zaman içindeki değişimini gösteren çizgi grafik",xlabel="Tarih",ylabel="Satış miktarı")
plt.show()
```

## bar grafiğinde aylık yapılan satışları göstermek 

```python
aylıksatıs=df.resample("ME")["Satış"].sum()
aylıksatıs.plot(kind="bar",title="aylık toplam satışlar",xlabel="ay",ylabel="toplam satış miktarı")
plt.show()
```

## Kategorilere göre satış dagılımı psata grafik 

Excel dosyandaki veriyi okur.
Kategori sütununa göre gruplar ve her kategorinin toplam satışını hesaplar.
plt.pie() ile pasta grafiğini çizer.
labels=kategori_satis.index → kategori isimlerini etiketler.
autopct="%1.1f%%" → yüzde değerlerini gösterir.
startangle=90 → grafiği 90 dereceden başlatır, daha düzenli görünür.

```python
kategori_satis = df.groupby("Kategori")["Satış"].sum()

plt.figure(figsize=(5, 5))
plt.pie(kategori_satis, labels=kategori_satis.index, autopct="%1.1f%%", startangle=90)
plt.title("Kategorilere Göre Satış Dağılımı")
plt.show()
```

 ## scatter plot (dağılım grafiği)

```python
df.plot(
    kind="scatter",
    x="Fiyat (TL)", 
    y="Satış", 
    title="Fiyat ve Satış arasındaki dağılım"
)
plt.show()
```

## fiyat histogram grafiği

```python
df["Fiyat (TL)"].plot(kind="hist", bins=10, title="Fiyat Dağılımı")
plt.xlabel("Fiyat (TL)")
plt.ylabel("Frekans")
plt.show()
```

 bins=10 → histogramı 10 aralığa böler (istersen arttırıp azaltabilirsin).
 xlabel → X ekseni etiketi (fiyat aralıkları).
 ylabel → Y ekseni etiketi (kaç ürün bu aralıkta).
